# Looking for instances of imported packages

```python
"""
Purpose: Want to find which functions from a certain package
are used within a file

Pseudocode:
1) define the model prefi or function names
2) Put them all into or regex statement
3) parse the file for 
"""
```

```python
from python_tools import file_utils as filu 
from pathlib import Path

filepaths = list(Path("/neurd_packages/mesh_tools/mesh_tools/").iterdir())
word_pat = "[a-zA-Z1-9_-]"
filu.file_regex_search(
    filepath = filepaths,
    patterns = [
        f"nviz.({word_pat}*)\(",
        f"cu\.({word_pat}*)\(",
        
        f"sm\.({word_pat}*)\(",
        f"soma_utils.({word_pat}*)\(",
    ],
    verbose = True,
    unique = True,
)
```
